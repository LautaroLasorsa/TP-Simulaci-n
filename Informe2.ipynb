```python
from simulacion2 import Simulacion
import matplotlib.pyplot as plt
import math
import threading
from multiprocessing import Process, Queue
```

```python
# Ejemplo de ejecución de la simulación
N = 100  # Número de personas
M = 10   # Número de entidades
tiempo = 1000
til_base = 5
tic_base = 7
memoria = 0.5
max_tolerancia_riesgo = 1.0
umbral = 1.5
beta = 10

simulacion = Simulacion(N, M, til_base, tic_base, memoria, max_tolerancia_riesgo, umbral, fun_contagio = lambda contagios, asistencias: 1-math.exp(-contagios/(beta*asistencias)))
historial_infectados, historial_satisfaccion, historial_asistencias = simulacion.ejecutar_simulacion(tiempo)
```

```python
plt.plot(historial_infectados, label='Infectados')
plt.plot(historial_satisfaccion, label='Satisfacción')
plt.plot([x/M for x in historial_asistencias], label='Asistencias')
plt.legend()
plt.show()
m_infectados = max(historial_infectados)
plt.plot([x/m_infectados for x in historial_infectados], label = "Infectados reescalados")
plt.show()
```

## Informe Paralelizado

```python
N = 1000  # Number of people
M = 10   # Number of entities
tiempo = 500
til_base = 2
tic_base = 2
memoria = 0.5
#max_tolerancia_riesgo = 0.0001
umbral = 1.5
beta = 10

def Caso(max_tolerancia_riesgo, queue):
    simulacion = Simulacion(
        N, M, til_base, tic_base, memoria, max_tolerancia_riesgo, umbral,
        fun_contagio=lambda contagios, asistencias: 1 - math.exp(-contagios / (beta * asistencias)),
        initial_exposeds=10
    )
    historial_infectados, historial_satisfaccion, historial_asistencias = simulacion.ejecutar_simulacion(tiempo)
    queue.put((max_tolerancia_riesgo, historial_infectados, historial_satisfaccion, historial_asistencias))

max_tolerancia_riesgos = [1e99, 1, 0.01, 0.001, 0.0005, 0.000375, 0.00025, 0.0001]
```

```python
import time 

def run_parallel():
    queue = Queue()
    processes = []

    for max_tolerancia_riesgo in max_tolerancia_riesgos:
        process = Process(target=Caso, args=(max_tolerancia_riesgo, queue))
        processes.append(process)

    start_time = time.time()
    for process in processes:
        process.start()

    for process in processes:
        process.join()
    end_time = time.time()

    results = [queue.get() for _ in processes]
    print(f"Parallel execution time: {end_time - start_time:.2f} seconds")
    return results

def run_sequential():
    results = []
    start_time = time.time()
    for max_tolerancia_riesgo in max_tolerancia_riesgos:
        queue = Queue()
        Caso(max_tolerancia_riesgo, queue)
        results.append(queue.get())
    end_time = time.time()

    print(f"Sequential execution time: {end_time - start_time:.2f} seconds")
    return results
```

```python
# results = run_parallel()

results_sequential = run_sequential()
```

```python
nrows, ncols = 4, 2
fig_size = (40, 24)
figA, axsA = plt.subplots(nrows, ncols, figsize = fig_size, sharex = True, sharey = True)
figB, axsB = plt.subplots(nrows, ncols, figsize = fig_size, sharex = True, sharey = True)

# Graficar los resultados
for i, (max_tolerancia_riesgo, historial_infectados, historial_satisfaccion, historial_asistencias) in enumerate(sorted(results_sequential, key = lambda x: -x[0])):
    y = i // ncols
    x = i % ncols

    axsA[y][x].plot(historial_infectados, label='Infected')
    axsA[y][x].plot(historial_satisfaccion, label='Satisfaction')
    axsA[y][x].plot([a / M for a in historial_asistencias], label='Attendance')
    axsA[y][x].legend()
    axsA[y][x].set_title(f"Max tolerance risk: {max_tolerancia_riesgo}", fontsize = 24)
  
    m_infectados = max(historial_infectados)
    
    axsB[y][x].plot([a / m_infectados for a in historial_infectados], label="Rescaled infected")
    axsB[y][x].set_title(f"Max tolerance risk: {max_tolerancia_riesgo}", fontsize = 24)
    
    if y == 3:
        axsA[y][x].set_xlabel('Time', fontsize = 36)
        axsB[y][x].set_xlabel('Time', fontsize = 36)
    
    if x == 0:
        axsA[y][x].set_ylabel('Count', fontsize = 36)
        axsB[y][x].set_ylabel('Count', fontsize = 36)
    
    axsA[y][x].legend()
    axsB[y][x].legend()

plt.tight_layout()
```
